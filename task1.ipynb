```python
import os
import numpy as np
import pandas as pd
```

```python
# Define the path to the folder containing the files
folder_path = r"C:\Users\Dell\Downloads\NLP-Tasks\multiple_sentences"

# Get a list of the files in the folder
files = os.listdir(folder_path)

# Create an empty DataFrame to store the file names
df = pd.DataFrame(columns=["File Name"])

# Loop through the files and add their names to the DataFrame
for file in files:
    df = df.append({"File Name": file}, ignore_index=True)

# Print the DataFrame
print(df)
```

```python
# Define function to remove first 5 characters and last 7 characters
def remove_chars(value):
    return value[5:9]

# Apply function to DataFrame column
df["clean_name"] = df["File Name"].apply(remove_chars)

print(df)
```

```python
# Drop "File Name" from DataFrame
df = df.drop("File Name", axis=1)
```

```python
# Rename "clean_name" to "New_ID"
df = df.rename(columns={"clean_name": "New_ID"})
```

```python
# Convert "ID" column in df2 to string data type
df["New_ID"] = df["New_ID"].astype(np.int64)
```

```python
# Read the Excel file into a Pandas DataFrame
df_metadata = pd.read_excel("BVC_Ethnicity.xlsx")

# Print the DataFrame to the console
print(df_metadata)
```

```python
# Merge DataFrames on "New_ID" column
output_df = pd.merge(df, df_metadata, on="New_ID", how="outer")

# Copy "Ethnicity" column to new column called "Original Ethnicity"
output_df["Original Ethnicity"] = output_df["Ethnicity"]

# Drop "Ethnicity" column
output_df = output_df.drop("Ethnicity", axis=1)

print(output_df)
```
